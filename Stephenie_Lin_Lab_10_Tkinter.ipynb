```python
from tkinter import *
import requests
import random

root = Tk()
root.geometry("300x200")
root.title("Quote Generator")

def get_authors():
    response = requests.get('https://api.quotable.io/authors')
    data = response.json()
    authors = [author['name'] for author in data['results']]
    return authors

def get_quote(author):
    response = requests.get(f'https://api.quotable.io/quotes?author={author}')
    data = response.json()
    quote = random.choice(data['results'])
    return quote['content']

def reset_menu():
    author_list = random.sample(get_authors(), 3)
    author_var.set(author_list[0])
    author_dropdown['menu'].delete(0, 'end')
    for author in author_list:
        #command=lambda allow us to send multiple data through the callback function.
        author_dropdown['menu'].add_command(label=author, command=lambda author=author: author_var.set(author))


author_label = Label(root, text="Choose an author:")
author_label.pack()
author_var = StringVar()
author_dropdown = OptionMenu(root, author_var, "")
author_dropdown.pack()

quote_button = Button(root, text="Get Quote", command=lambda: print(get_quote(author_var.get())))
quote_button.pack()

reset_button = Button(root, text="Reset Authors", command=reset_menu)
reset_button.pack()

reset_menu()

root.mainloop()
```
