```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv

PATH = Service("/Users/stephenielin/Downloads/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

driver.get("https://www.binghamton-ny.gov/home")

govt = '//*[@id="dropdownrootitem3"]'
dept = '//*[@id="dropdownrootitem3"]/div/div/ul[1]/li/a'

time.sleep(5)
govt_tab = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, govt)))
time.sleep(5)
ActionChains(driver).move_to_element(govt_tab).click().perform()
time.sleep(5)
dept_link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, dept)))
time.sleep(5)
ActionChains(driver).move_to_element(dept_link).click().perform()

try:
    personnel = '//*[@id="widget_4_33_127"]/ul/li[16]/a'
    employment = '//*[@id="leftNav_1038_0_145"]/ul/li/ul/li[14]/ul/li/a'
    personnel_link = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, personnel)))
    personnel_link.click()
    employment_link = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, employment)))
    employment_link.click()

    table = driver.find_element(By.XPATH, '//*[@id="ColumnUserControl4"]/div[2]')
    rows = table.find_elements(By.TAG_NAME, "tr")
    with open('job_openings.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Job", "Salary", "Deadline", "Job Type"])
        for i in rows[1:]: 
            job = i.find_element(By.TAG_NAME, "td").text
            job_type = i.find_element(By.XPATH, 'td[2]').text
            deadline = i.find_element(By.XPATH, 'td[3]').text
            salary = i.find_element(By.XPATH, 'td[4]').text
            writer.writerow([job, salary, deadline, job_type])

finally:
    driver.quit()

import pandas as pd
df = pd.read_csv('job_openings.csv')
print(df)


```
