```python
import csv
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

try:
    driver.get("https://www.wikipedia.org/")
    search = driver.find_element(By.NAME, "search")
    search.send_keys("cat")
    search.submit()

    #https://www.guru99.com/xpath-selenium.html, By.XPATH is used to find elements not able to be easily found
    cat_link = driver.find_element(By.XPATH, "//a[@href='/wiki/Cat']")
    cat_link.click()

    img_link = []
    images = driver.find_elements(By.XPATH, "//div[@class='thumbinner']//a[@class='image']//img")
    for image in images:
        try:
            image_link = image.find_element(By.XPATH, "..").get_attribute("href")
        except:
            driver.quit()
            
        if image_link:
            img_link.append(image_link)

    with open("img_link.csv", "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows([[link] for link in img_link])

    # Read the CSV file in a pandas DataFrame, so it's easier to read
    df = pd.read_csv("img_link.csv", header=None, names=["img_link"])
    print(df)

finally:
    driver.quit()

```

```python
#bonus
#scrape all <h3> heading
import csv
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

try:
    driver.get("https://www.wikipedia.org/")
    search = driver.find_element(By.NAME, "search")
    search.send_keys("cat")
    search.submit()

    h3_list = driver.find_elements(By.TAG_NAME, "h3")
    h3_headlines = []
    for h3 in h3_list:
        try:
            h3_text = h3.text
        except:
            driver.quit()
        if h3_text:
            h3_headlines.append(h3_text)

    with open("h3_headlines.csv", "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Headlines"])
        for headline in h3_headlines:
            writer.writerow([headline])

    # Read the CSV file in a pandas DataFrame, so it's easier to read
    df = pd.read_csv("h3_headlines.csv")
    print(df)

finally:
    driver.quit()
```
